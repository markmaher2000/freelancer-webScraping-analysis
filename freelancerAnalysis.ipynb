### What percentage of freelancers complete projects on time and within budget?

```python
time_budget = df.query("On_Time == 100 and On_Budget == 100")
countThem = len(time_budget)
thePercentage = (countThem / len(df)) * 100
```

```python
print(f"The Percentage of freelancers complete projects on time and on budget {round(thePercentage, 2)}%")
```

### Let's check for each skill

```python
l_skill = list(time_budget['Skill'].unique())
l_skill
```

```python
l_percentage = []
for i in l_skill:
    skill = time_budget.query(f"Skill ==  '{i}'")
    Total_ = df.query(f"Skill == '{i}'")
    percentageSkill = (len(skill) / len(Total_)) * 100
    print(f"The Percentage of freelancers {i} complete projects on time and on budget {round(percentageSkill, 2)}%")
    l_percentage.append(percentageSkill)


```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x=l_skill, y=l_percentage)
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.2f}%',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, fontweight='bold', color='Black'
    )
plt.xlabel("Skill")
plt.ylabel("Percentage %")
plt.xticks(rotation=30)
plt.title("Bar Plot for freelancers complete projects on time and on budget According to each Skill")
```

### How does hourly pay vary across different skills?

```python
avg_pay_by_skill = df.groupby("Skill")["Hourly_Paid"].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,5))
ax = sns.barplot(data=avg_pay_by_skill, x="Skill", y="Hourly_Paid", hue="Skill")
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.2f}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, fontweight='bold', color='Black'
    )
plt.xlabel("Skill")
plt.ylabel("Hourly Paid")
plt.xticks(rotation=45)
plt.show();
```

```python
work_df = df.query("Reviews > 0")
avg_pay_by_skill = work_df.groupby("Skill")["Hourly_Paid"].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,5))
ax = sns.barplot(data=avg_pay_by_skill, x="Skill", y="Hourly_Paid", hue="Skill")
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.2f}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, fontweight='bold', color='Black'
    )
plt.xlabel("Skill")
plt.ylabel("Hourly Paid")
plt.xticks(rotation=45)
plt.show();
```
